```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

QNo = 3
```

```python
# Data path is of form "output/[output Q Number]/[cnn]/[config]/"
data_path = "./output/outputQ"+str(QNo)+"/"
save_path = "./graph/Q"+str(QNo)+"/"

# Get all CNNs
cnn_list = os.listdir(data_path)
cnn_list = [cnn for cnn in cnn_list if os.path.isdir(data_path + cnn)]

# Get all configurations
config_list = os.listdir(data_path + cnn_list[0])
config_list = [config for config in config_list if os.path.isdir(data_path + cnn_list[0] + "/" + config)]

# folder_list = []

# for cnn in cnn_list:
#     for config in config_list:
#         folder_list.append(cnn + "/" + config)

# Get all output files
compute_files = []
bandwidth_files = []
detailed_access_files = []


for cnn in cnn_list:
    for config in config_list:
        compute_files.append(data_path + cnn + "/"+ config + "/COMPUTE_REPORT.csv")
        
        
# Read all output files
compute = []

for file in compute_files:
    compute.append(pd.read_csv(file))
    
# Do the same for bandwidth files


for cnn in cnn_list:
    for config in config_list:
        bandwidth_files.append(data_path + cnn + "/" + config + "/BANDWIDTH_REPORT.csv")
        
# Read all output files
bandwidth = []

for file in bandwidth_files:
    bandwidth.append(pd.read_csv(file))
    
# Do the same for detailed access files


for cnn in cnn_list:
    for config in config_list:
        detailed_access_files.append(data_path + cnn + "/" + config + "/DETAILED_ACCESS_REPORT.csv")
        
# Read all output files
detailed_access = []

for file in detailed_access_files:
    detailed_access.append(pd.read_csv(file))
```

```python
# Remove last column
for i in range(len(compute)):
    compute[i] = compute[i].iloc[:, :-1]
compute[0].head()
```

```python
# Remove last column
for i in range(len(bandwidth)):
    bandwidth[i] = bandwidth[i].iloc[:, :-1]
bandwidth[0].head()
```

```python
# Remove last column from detailed access
for i in range(len(detailed_access)):
    detailed_access[i] = detailed_access[i].iloc[:, :-1]
    
detailed_access[0].head()
```

```python
# CNN number is i, configuration number is j
# Thus i*len(config_list) + j is the index of the dataframes

# Plot the Mapping Efficiency %

# Get the mapping efficiency
mapping_efficiency = []

for i in range(len(cnn_list)):
    for j in range(len(config_list)):
        mapping_efficiency.append(compute[i*len(config_list) + j][" Mapping Efficiency %"])

# Plot the mapping efficiency
for i in range(len(cnn_list)):
    fig, ax = plt.subplots(1, 1, figsize=(15, 5))
    for j in range(len(config_list)):
        layer = compute[i*len(config_list) + j]["LayerID"]
        # Reverse bar graph
        ax.plot(layer, mapping_efficiency[i*len(config_list) + j], label=config_list[j])
    ax.set_title(cnn_list[i])
    ax.set_xlabel("Layer")
    ax.set_ylabel("Mapping Efficiency %")
    ax.legend()
    plt.savefig(save_path+cnn_list[i]+"_mapping_efficiency.png")
```

```python
# Plot the Compute Util %

# Get the Compute Util
compute_efficiency = []

for i in range(len(cnn_list)):
    for j in range(len(config_list)):
        compute_efficiency.append(compute[i*len(config_list) + j][" Compute Util %"])
        
# Plot the Compute Util
for i in range(len(cnn_list)):
    fig, ax = plt.subplots(1, 1, figsize=(15, 5))
    for j in range(len(config_list)):
        layer = compute[i*len(config_list) + j]["LayerID"]
        ax.plot(layer, compute_efficiency[i*len(config_list) + j], label=config_list[j])
    ax.set_title(cnn_list[i])
    ax.set_xlabel("Layer")
    ax.set_ylabel("Compute Util %")
    ax.legend()
    plt.savefig(save_path+cnn_list[i]+"_compute_util.png")
```

```python
# Total Cycle Count

# Get the Total Cycle Count
total_cycle_count = []

for i in range(len(cnn_list)):
    for j in range(len(config_list)):
        total_cycle_count.append(compute[i*len(config_list) + j][" Total Cycles"])
        
# Plot the Total Cycle Count

for i in range(len(cnn_list)):
    fig, ax = plt.subplots(1, 1, figsize=(15, 5))
    for j in range(len(config_list)):
        layer = compute[i*len(config_list) + j]["LayerID"]
        ax.plot(layer, total_cycle_count[i*len(config_list) + j], label=config_list[j])
    ax.set_title(cnn_list[i])
    ax.set_xlabel("Layer")
    ax.set_ylabel("Total Cycles")
    ax.legend()
    plt.savefig(save_path+cnn_list[i]+"_total_cycles.png")
```

```python
# Bandwidth stuff

para = [" Avg IFMAP SRAM BW"," Avg FILTER SRAM BW"," Avg OFMAP SRAM BW"," Avg IFMAP DRAM BW"," Avg FILTER DRAM BW"," Avg OFMAP DRAM BW"]

for p in para:
    # Get the bandwidth
    bandwidth_para = []

    for i in range(len(cnn_list)):
        for j in range(len(config_list)):
            bandwidth_para.append(bandwidth[i*len(config_list) + j][p])

    # Plot the bandwidth
    for i in range(len(cnn_list)):
        fig, ax = plt.subplots(1, 1, figsize=(15, 5))
        for j in range(len(config_list)):
            layer = bandwidth[i*len(config_list) + j]["LayerID"]
            ax.plot(layer, bandwidth_para[i*len(config_list) + j], label=config_list[j])
        ax.set_title(cnn_list[i])
        ax.set_xlabel("Layer")
        ax.set_ylabel(p)
        ax.legend()
        plt.savefig(save_path+cnn_list[i]+"_"+p+".png")
```

```python
# Reads and Writes

para = [" SRAM IFMAP Reads", " SRAM Filter Reads", " SRAM OFMAP Writes", " DRAM IFMAP Reads", " DRAM Filter Reads", " DRAM OFMAP Writes"]

for p in para:
    # Get the detailed access
    detailed_access_para = []

    for i in range(len(cnn_list)):
        for j in range(len(config_list)):
            detailed_access_para.append(detailed_access[i*len(config_list) + j][p])

    # Plot the detailed access
    for i in range(len(cnn_list)):
        fig, ax = plt.subplots(1, 1, figsize=(15, 5))
        for j in range(len(config_list)):
            layer = detailed_access[i*len(config_list) + j]["LayerID"]
            ax.plot(layer, detailed_access_para[i*len(config_list) + j], label=config_list[j])
        ax.set_title(cnn_list[i])
        ax.set_xlabel("Layer")
        ax.set_ylabel(p)
        ax.legend()
        plt.savefig(save_path+cnn_list[i]+"_"+p+".png")
```
